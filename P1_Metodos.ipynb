# Modelo mais simples **(SIER)** ü¶†

Primeiramente, vamos explicar o modelo epidemiol√≥gico SIR, proposto por Kermack e McKendrick em 1927, √© a base dos modelos epidemiol√≥gicos mais sofisticados que conhecemos hoje. Ele √© baseado na hip√≥tese de que o indiv√≠duo de uma determinada popula√ß√£o, onde ocorre uma epidemia, passa por diferentes est√°gios de suscetibilidade √† infec√ß√£o. 



Tomando como premissa de que a popula√ß√£o √© constante e as *taxas de crescimento* e *remo√ß√£o da doen√ßa* s√£o denotadas por Œ≤ e Œ≥, respectivamente. Isso leva √†s equa√ß√µes:

\begin{align}
\frac{{\rm d}S}{{\rm d}t} & = -\beta S I/N\\
\frac{{\rm d}I}{{\rm d}t} & = \beta S I/N- \gamma I\\
\frac{{\rm d}R}{{\rm d}t} & = \gamma I\\
\end{align}

Onde 1 / Œ≥ √© o per√≠odo m√©dio de tempo durante o qual um indiv√≠duo infectado pode transmitir o v√≠rus.

Podemos observar que a soma total dS/dt + dI/dt + dR/dt √© igual a zero, j√° que a popula√ß√£o √© considerada constante. Al√©m disso, analisando cada termo, podemos notar que a taxa de varia√ß√£o dos indiv√≠duos suscet√≠veis depende do contato do grupo S com o grupo I, o que √© relevante para as an√°lises futuras.

#### Para o valor de Œ≥:


Œ≤ e Œ≥ s√£o valores utilizados tanto para o modelo SIR como para o SIDR. O c√°lculo do valor pra Œ≥ √© o seguinte:

Œ≥ = 1/dias infectados

Œ≥ = 1/15

Obs: 
1. **Suscet√≠veis S(t):**¬†pessoas que est√£o suscet√≠veis a infec√ß√£o
2. **Infectados¬†I(t):**¬†pessoas que est√£o infectadas com o v√≠rus

4. **Recuperados¬†R(t):**¬†pessoas que se recuperaram da doen√ßa ou se tornaram imunes atrav√©s da vacina√ß√£o


#### C√°lculo de Œ≤ a partir de Rt0:

Como, Rt0 = Œ≤/Œ≥:

√â necess√°rio encontrar o n√∫mero b√°sico de reprodu√ß√£o da infec√ß√£o(Rt0), que mede a infectividade de um pat√≥geno em um ambiente no qual ningu√©m adquiriu imunidade a ele, √© um valor essencial para a avalia√ß√£o da situa√ß√£o, pois: A

Se Rt0 < 1: cada infec√ß√£o existente causa menos do que uma nova infec√ß√£o. A doen√ßa diminuir√° e eventualmente desaparecer√°
Se Rt0 = 1: cada infec√ß√£o existente causa uma nova infec√ß√£o. A doen√ßa permanecer√° viva e est√°vel, mas n√£o haver√° um surto ou uma epidemia
Se Rt0 > 1: cada infec√ß√£o existente causa mais de uma nova infec√ß√£o. A doen√ßa √© transmitida entre as pessoas e pode haver um surto ou epidemia.
O Rt0 do v√≠rus Sars-CoV-2, causador da Covid-19, est√° entre 2,5 e 3‚Äâ‚Äì‚Äâo que significa que um infectado contamina, em m√©dia, duas ou tr√™s pessoas.

O Rt0 √© calculado com base em tr√™s informa√ß√µes: o n√∫mero de contatos que uma pessoa infectada faz com indiv√≠duos suscet√≠veis, o risco de transmiss√£o em cada contato realizado e o tempo m√©dio em que o infectado transmite a doen√ßa. Na Covid-19, a transmiss√£o come√ßa cerca de dois dias antes do doente apresentar sintomas e se prorroga por mais sete.

A partir de Rt0, podemos calcular o valor para Rt(n√∫mero efetivo de reprodu√ß√£o), que √© o n√∫mero m√©dio de indiv√≠duos contagiados por cada infectado nas condi√ß√µes existentes em um determinado momento. O seu resultado poder√° classificar uma doen√ßa como end√™mica, epid√™mica ou em queda.

Rt = s . Rt0

com S = propor√ß√£o de suscet√≠veis da popula√ß√£o

Caso tenhamos uma propor√ß√£o de suscet√≠veis = 80% e um Rt0 de 2:

Rt = 0.8 * 2

Rt = 1.6

Rt > 1: O n√∫mero de casos da doen√ßa est√° aumentando. Epidemia
Rt = 1: Cada infectado causa uma nova infec√ß√£o. Endemia
Rt < 1: A doen√ßa diminuir√° e eventualmente acabar√°.
Ent√£o, no exemplo, a situa√ß√£o √© epid√™mica. Por√©m para acharmos Œ≤, n√£o ser√° necess√°rio o Rt, apenas o Rt0:

a partir dos valores conhecidos de Rt0 e Œ≥, podemos achar Œ≤ usando Rt0 = Œ≤/Œ≥

#### SIER

O modelo que utilizaremos neste projeto nada mais √© do que uma varia√ß√£o do modelo SIR, que leva em considera√ß√£o mais vari√°veis para aproximar mais o resultado da realidade. 

Antes de aplicarmos o modelo proposto, iremos mostrar os resultados do modelo SEIR, para, no fim, compar√°-lo com o SIDR. 



Para o modelo SEIR, precisamos adicionar uma nova equa√ß√£o para o grupo E. Assumindo que Œ¥ √© o tempo que um indiv√≠duo infectado leva para se tornar infeccioso, as equa√ß√µes s√£o:







\begin{align}
\frac{{\rm d}S}{{\rm d}t} & = -\beta S (I+qE)/N\\
\frac{{\rm d}E}{{\rm d}t} & = \beta S (I+qE)/N- E/\delta \\
\frac{{\rm d}I}{{\rm d}t} & = E/\delta-\gamma I\\
\frac{{\rm d}R}{{\rm d}t} & = \gamma I\\
\end{align}

Observe que o grupo E aparece na equa√ß√£o do grupo S, pois um indiv√≠duo exposto pode espalhar o v√≠rus sem demonstrar sintomas. A constante `q` √© usada para contabilizar a menor probabilidade de um indiv√≠duo exposto infectar outra pessoa, comparado √† probabilidade de um indiv√≠duo infeccioso fazer o mesmo.

Para aprimorar o modelo SEIR, √© necess√°rio acrescentar novas vari√°veis, como nosso objetivo √© analisar o impacto do SJ no n√∫mero de pessoas com covid, iremos adicionar vari√°veis relacionadas a taxa de vacinacao, e iremos mudar a premissa de popula√ß√£o constante.

# Modelo com mais vari√°veis **(SIDR)** ü¶†

No modelo SIDR iremos adicionar as vari√°veis de √≥bitos e vacina√ß√£o.

Iremos nos basear em um modelo amplamente conhecido chamado de *SIDR*.
Para discretizar esse problema, vamos reutilizar alguns temos do t√≥pico anterior e modific√°los.

Como j√° era esperado, a popula√ß√£o n√£o ser√° constante aqui, ela ser√° vari√°vel em virtude do n√∫mero de √≥bitos que a covid-19 causa. Para facilitar o entendimento, a popula√ß√£o ser√° dividida da seguinte forma:

- **S(t):** pessoas suscet√≠veis a pegar a doenca.
- **I(t):** pessoas infectadas pelo v√≠rus.
- **D(t):** Pessoas que morreram devido ao virus.
- **R(t):** pessoas que se recuperaram ou foram imunizadas pela vacina.









Para fazer a modelagem, utilizaremos os seguintes par√¢metros:

œà ‚Üí	Representa a taxa de vacina√ß√£o e a efic√°cia de vacina

Œ≤ ‚Üí	Representa a taxa de infec√ß√£o do v√≠rus 

Œ≥ ‚Üí	Representa a taxa de recupera√ß√£o

œÅ ‚Üí	Representa a taxa de mortalidade em decorr√™ncia da doen√ßa

Œ¶ ‚Üí	Representa a taxa de imuniza√ß√£o  


A equa√ß√£o dS(t)/dt descreve a taxa de varia√ß√£o ao longo do tempo do grupo de indiv√≠duos suscet√≠veis √† infec√ß√£o, levando em considera√ß√£o a taxa de transmiss√£o Œ≤, o n√∫mero de indiv√≠duos infectados I(t), a taxa de recupera√ß√£o Œ≥, a taxa de mortalidade œÅ, a taxa de imuniza√ß√£o Œ¶ e a taxa de vacina√ß√£o e efic√°cia da vacina œà.


\begin{align}
 \frac {dS(t)} {dt} = - Œ≤.S(t).I(t) + Œ≥.I(t).(1 - Œ¶) - S(t).œà
\end{align}
\begin{align}
\frac {dI(t)} {dt} = Œ≤.S(t).I(t) - (Œ≥/1-œÅ).I(t)
\end{align}
\begin{align}
\frac {dD(t)} {dt} = (œÅ/1-œÅ).Œ≥.I(t)
\end{align}
\begin{align}
\frac {dR(t)} {dt} = Œ≥.I(t).Œ¶ + S(t).v
\end{align}

De maneira an√°loga, ao modelo SIR, teremos: *N = S(t) + I(t) + D(t) + R(t)*
