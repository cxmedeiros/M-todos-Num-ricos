# **Questão *10b*** - lista 2 - Métodos numéricos

**Importando as bibliotecas necessárias**

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

**definindo o modelo**

```python
def model(y, t):
    y0, y1 = y
    dydt = [y1, -y0 + np.cos(t)]
    return dydt
```

**definidno a condição inicial, o intervalo e resolvendo a equação**

```python
y0 = [0, 0] # initial conditions
t = np.linspace(0, 10*np.pi, 1000) # time points to solve the equation
y = odeint(model, y0, t)
```

**Solução Analítica**

```python
y_analitico = t*np.sin(t)/2
```

**Plotando a solução**

```python
plt.plot(t, y[:,0], 'b', label='y(t)') # plot the solution
plt.plot(t, y_analitico, 'r--', color= "red" )
plt.xlabel('t')
plt.ylabel('y')
plt.legend(['Solução Numérica', 'Solução Analítica'])
plt.show()
```
