#**(SIR)** ü¶†

Primeiramente, vamos explicar o modelo epidemiol√≥gico SIR, proposto por Kermack e McKendrick em 1927, √© a base dos modelos epidemiol√≥gicos mais sofisticados que conhecemos hoje. Ele √© baseado na hip√≥tese de que o indiv√≠duo de uma determinada popula√ß√£o, onde ocorre uma epidemia, passa por diferentes est√°gios de suscetibilidade √† infec√ß√£o.

O sistema SIR √© fundamentado em tr√™s grupos:



- **S(t):** pessoas suscet√≠veis a pegar a doen√ßa.
- **I(t):** pessoas infectadas pelo v√≠rus.
- **R(t):** pessoas que se recuperaram ou foram imunizadas pela vacina.



Sob a premissa de uma popula√ß√£o invari√°vel, onde a taxa de propaga√ß√£o da doen√ßa √© denotada por  Œ≤  e a taxa m√©dia de recupera√ß√£o √© representada por  Œ≥ , podemos expressar as equa√ß√µes da seguinte forma:

\begin{align}
\frac{{\rm d}S}{{\rm d}t} & = -\beta S I/N\\
\frac{{\rm d}I}{{\rm d}t} & = \beta S I/N- \gamma I\\
\frac{{\rm d}R}{{\rm d}t} & = \gamma I\\
\end{align}

Onde 1 / Œ≥ √© o per√≠odo m√©dio de tempo durante o qual um indiv√≠duo infectado pode transmitir o v√≠rus.

Podemos observar que a soma total dS/dt + dI/dt + dR/dt √© igual a zero, j√° que a popula√ß√£o √© considerada constante. Al√©m disso, analisando cada termo, podemos notar que a taxa de varia√ß√£o dos indiv√≠duos suscet√≠veis depende do contato do grupo S com o grupo I, o que √© relevante para as an√°lises futuras.

#**(SIER)** ü¶†

Este nada mais √© do que uma varia√ß√£o do modelo SIR, que leva em considera√ß√£o mais vari√°veis para aproximar mais o resultado da realidade. 

Antes de aplicarmos o modelo proposto, iremos mostrar os resultados do modelo SEIR, para, no fim, compar√°-lo com o SIDR. 



Para o modelo SEIR, precisamos adicionar uma nova equa√ß√£o para o grupo E.

**Expostos E (t):**¬†pessoas que est√£o expostas ao v√≠rus.

 Assumindo que Œ¥ √© o tempo que um indiv√≠duo infectado leva para se tornar infeccioso, as equa√ß√µes s√£o:







\begin{align}
\frac{{\rm d}S}{{\rm d}t} & = -\beta S (I+qE)/N\\
\frac{{\rm d}E}{{\rm d}t} & = \beta S (I+qE)/N- E/\delta \\
\frac{{\rm d}I}{{\rm d}t} & = E/\delta-\gamma I\\
\frac{{\rm d}R}{{\rm d}t} & = \gamma I\\
\end{align}

Observe que o grupo E aparece na equa√ß√£o do grupo S, pois um indiv√≠duo exposto pode espalhar o v√≠rus sem demonstrar sintomas. A constante `q` √© usada para contabilizar a menor probabilidade de um indiv√≠duo exposto infectar outra pessoa, comparado √† probabilidade de um indiv√≠duo infeccioso fazer o mesmo.

Para aprimorar o modelo SEIR, √© necess√°rio acrescentar novas vari√°veis, como nosso objetivo √© analisar o impacto do SJ no n√∫mero de pessoas com covid, iremos adicionar vari√°veis relacionadas a taxa de vacinacao, e iremos mudar a premissa de popula√ß√£o constante.

# **(SIDR)** ü¶†

No modelo SIDR iremos adicionar as vari√°veis de √≥bitos e vacina√ß√£o.

Œ¶ ‚Üí	Representa a taxa de imuniza√ß√£o  


A equa√ß√£o dS(t)/dt descreve a taxa de varia√ß√£o ao longo do tempo do grupo de indiv√≠duos suscet√≠veis √† infec√ß√£o, levando em considera√ß√£o a taxa de transmiss√£o Œ≤, o n√∫mero de indiv√≠duos infectados I(t), a taxa de recupera√ß√£o Œ≥, a taxa de mortalidade œÅ, a taxa de imuniza√ß√£o Œ¶ e a taxa de vacina√ß√£o e efic√°cia da vacina œà.


\begin{align}
 \frac {dS(t)} {dt} = - Œ≤.S(t).I(t)
\end{align}
\begin{align}
\frac {dI(t)} {dt} = Œ≤.S(t).I(t) - (Œ≥/1-œÅ).I(t)

\begin{align}

# Aplicando Modelo **SIR**ü¶†

```python
import numpy as np
from scipy.integrate import odeint
import plotly.express as px
import pandas as pd
```

```python
N = 911329 + 21506 + 766013 # confirmados + obitos + recuperados de 12 de abril de 2022

# Numero Inicial de Infectados, Removidos e Mortos
I0 = int(911329 / N * 100)

R0 = int(766013 / N * 100)
print(I0, R0)
# Todas as outras pessoas s√£o suscet√≠veis, inicialmente
S0 = N - I0 - R0

# Taxa de Infec√ß√£o
# Taxa de Recupera√ß√£o, 1/dias infectados(m√©dia de 15)
gamma = 1/15
beta = 0.5 # baseada na refer√™ncia
print(gamma)
print(beta)
# Dias Decorridos
t = np.linspace(0, 120, 120) 

def calcSIR(x, t):
  s,i,r = x
  dt = np.zeros(3)
  
  dt[0] = -beta * s*i/N 
  dt[1] = (beta*s*i/N) - gamma*i
  dt[2] = gamma*i

  return dt
dados = [S0, I0, R0]
x = odeint(calcSIR, dados, t)
s = x[:,0]
i = x[:,1]
r = x[:,2]
results = {'time':t, 'sucetiveis':s/N, 'infecciosos':i/N, 'recuperados':r/N}
result = pd.DataFrame.from_dict(results)

#PLOTANDO O GRAFICO
fig = px.line(result, x = 'time', y=['sucetiveis', 'infecciosos', 'recuperados'])
fig.update_layout(xaxis_title = 'Time in days', yaxis_title = 'Percent', title = 'SIR aplicated on Pernambuco Covid-19 Data in 2021')
fig.show()
```

# Aplicando Modelo **SIER**ü¶†

```python
import numpy as np
import pandas as pd
import plotly.express as px
from scipy.integrate import odeint

N = 911329 + 21506 + 766013 # confirmados + obitos + recuperados de 12 de abril de 2022

# Numero Inicial de Infectados, Removidos e Mortos
I0 = int(911329 / N * 100)

R0 = int(766013 / N * 100)
print(I0, R0)

# Todas as outras pessoas s√£o suscet√≠veis, inicialmente

S0 = N - I0 - R0

# Taxa de Infec√ß√£o
# Taxa de Recupera√ß√£o, 1/dias infectados(m√©dia de 15)
gamma = 1/15
beta = 0.5
delta = 5  # per√≠odo de incuba√ß√£o
q = 0.1  # propor√ß√£o de contatos infecciosos para expostos

E0 = S0 * q # n√∫mero inicial de expostos

t = np.linspace(0, 120, 120)

def calcSIER(x, t):
    s, e, i, r = x
    dsdt = -beta * s * (i + q * e) / N
    dedt = beta * s * (i + q * e) / N - e / delta
    didt = e / delta - gamma * i
    drdt = gamma * i
    return [dsdt, dedt, didt, drdt]

dados = [S0, E0, I0, R0]
x = odeint(calcSIER, dados, t)
s = x[:, 0]
e = x[:, 1]
i = x[:, 2]
r = x[:, 3]

results = {'time': t, 'susceptible': s/N, 'exposed': e/N, 'infectious': i/N, 'recovered': r/N}
result = pd.DataFrame.from_dict(results)

fig = px.line(result, x='time', y=['susceptible', 'exposed', 'infectious', 'recovered'])
fig.update_layout(xaxis_title='Time (days)', yaxis_title='Fraction of population', title='SIER Model Simulation')
fig.show()
```
